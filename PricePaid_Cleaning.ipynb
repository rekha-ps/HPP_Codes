```python
#Import necessary libraries
import pandas as pd
import numpy as np
import re
```

```python
#Loading price paid datasets to a dataframe
price_paid_df = pd.read_csv(r'C:\Users\Rekha\Desktop\Dissertation Datasets\Price_Paid.csv')
price_paid_df.head()
```

```python
#adding column names
price_paid_df.columns = ['Transaction_unique_reference', 'Price_Paid', 'Transaction date', 'Postcode', 'Property_Type' , 'Old/New' , 'Freehold/Leasehold' , 'AddressLine_1' , 'AddressLine_2' , 'Street', 'Locality' , 'Town/City' , 'District', 'County', 'PPD_Category' , 'RecordStatus']
price_paid_df.head()
```

```python
price_paid_df.shape
```

```python
#dropping unwanted columns
columns_to_drop = ['Transaction_unique_reference','PPD_Category', 'RecordStatus']
price_paid_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
#removing duplicates from the dataframe
price_paid_df.drop_duplicates()
```

```python
# Convert 'Transaction date' column to datetime
price_paid_df['Transaction date'] = pd.to_datetime(price_paid_df['Transaction date'])
price_paid_df.head()
```

```python
# Defining the start and end dates for desired date range
start_date = pd.to_datetime('2013-01-01')
end_date = pd.to_datetime('2023-07-30')

# Filtering the df to include rows within the date range provided
price_paid_df = price_paid_df[(price_paid_df['Transaction date'] >= start_date) & (price_paid_df['Transaction date'] <= end_date)]
                          
# Display the filtered DataFrame
price_paid_df.head()
```

```python
#check all the counties in the dataset
price_paid_df['County'].unique()
```

```python
#selecting the counties in yorksire and the humber
selected_counties = ['SOUTH YORKSHIRE','NORTH YORKSHIRE', 'WEST YORKSHIRE', 'EAST RIDING OF YORKSHIRE', 'CITY OF KINGSTON UPON HULL', 'LINCOLNSHIRE', 'NORTH EAST LINCOLNSHIRE', 'NORTH LINCOLNSHIRE', 'YORK']

mask = price_paid_df['County'].isin(selected_counties)
price_paid_df = price_paid_df[mask]
price_paid_df
```

```python
price_paid_df.isnull().sum()
```

```python
# Calculate the percentage of missing values in the 'Postcode' column
total_rows = len(price_paid_df)
missing_postcode_count = price_paid_df['Postcode'].isnull().sum()
percentage_missing_postcode = (missing_postcode_count / total_rows) * 100

# Print the calculated percentage
print(f"Percentage of missing postcodes: {percentage_missing_postcode:.2f}%")
```

```python
#dropping the rows with empty postcode
price_paid_df = price_paid_df.dropna(subset=['Postcode'])
```

```python
unique_address_1 = price_paid_df['AddressLine_1'].unique()

print("Unique values in 'AddressLine_1' column:")
for value in unique_address_1:
    print(value)
```

```python
unique_address_2 = price_paid_df['AddressLine_2'].unique()

print("Unique values in 'AddressLine_2' column:")
for value in unique_address_2:
    print(value)
```

```python
price_paid_df[['AddressLine_1','AddressLine_2','Street']].sample(10)
```

```python
columns_to_clean = ['AddressLine_1', 'AddressLine_2', 'Street']  

for column in columns_to_clean:
    price_paid_df[column] = price_paid_df[column].str.strip().str.title().str.replace(',', '')
```

```python
# Fill missing values in Street and AddressLine_2 with an empty string
price_paid_df[['Street', 'AddressLine_2']] = price_paid_df[['Street', 'AddressLine_2']].fillna('')

# Create a new Address column based on the requirements
price_paid_df['Address'] = price_paid_df.apply(lambda row:
    row['AddressLine_2'] + ' ' + row['AddressLine_1'] + ' ' + row['Street'] if row['AddressLine_2'].isnumeric()
    else row['AddressLine_1'] + ' ' + row['Street'] + ' ' + row['AddressLine_2'],
    axis=1
)

# Remove any extra spaces from the beginning or end of the Address column
price_paid_df['Address'] = price_paid_df['Address'].str.strip()

# Display the result
price_paid_df.head(10)
```

```python
# Display rows that contain "Floor" in the 'Address' column
filtered_rows = price_paid_df[price_paid_df['Address'].str.contains('Floor', case=False, na=False)]

# Print the filtered DataFrame
filtered_rows[['Address','Postcode']].sample(50)
```

```python
columns_to_drop = ['AddressLine_1','AddressLine_2','Street', 'Locality']
price_paid_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# Reorder columns in price_paid_df
price_paid_df = price_paid_df[['Transaction date', 'Price_Paid','Address', 'Town/City','Postcode','Property_Type','Old/New','Freehold/Leasehold','District','County']]
price_paid_df.head()
```

```python
# Rename columns in price_paid_df
price_paid_df.rename(columns={'Transaction date': 'Transaction_date'}, inplace=True)

# Print the updated DataFrame
price_paid_df.info()
```

```python
price_paid_df['Town/City'] = price_paid_df['Town/City'].str.title().str.strip()
price_paid_df['District'] = price_paid_df['District'].str.title().str.strip()
price_paid_df['County'] = price_paid_df['County'].str.title().str.strip()
price_paid_df['Address'] = price_paid_df['Address'].str.title().str.strip()
```

```python
csv_file_path = r'C:\Users\Rekha\Desktop\Dissertation Datasets\PricePaid_Cleaned.csv'

# Save the cleaned DataFrame to a CSV file
price_paid_df.to_csv(csv_file_path, index=False)
```
